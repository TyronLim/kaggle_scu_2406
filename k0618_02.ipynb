```python
fill1 = fill_tsh = df["TSH"].mean()
fill2 = fill_free_t3 = df["FreeT3"].mean()
fill3 = fill_free_t4 = df["FreeT4"].mean()
fill4 = fill_age = df['나이'].mean()
fill1, fill2, fill3, fill4
```

```python
# train
df["나이"] = df["나이"].fillna(fill4)
df["성별"] = df["성별"].fillna("여")
df["TSH"] = df["TSH"].fillna(fill_tsh)
df["FreeT3"] = df["FreeT3"].fillna(fill_free_t3)
df["FreeT4"] = df["FreeT4"].fillna(fill_free_t4)
```

```python
# test
df_t["나이"] = df_t["나이"].fillna(fill4)
df_t["성별"] = df_t["성별"].fillna("여")
df_t["TSH"] = df_t["TSH"].fillna(fill_tsh)
df_t["FreeT3"] = df_t["FreeT3"].fillna(fill_free_t3)
df_t["FreeT4"] = df_t["FreeT4"].fillna(fill_free_t4)
```

```python
# train
df['성별'] = df['성별'].replace('UNK', '여')
df['성별'].value_counts()
```

```python
# test
df_t['성별'] = df_t['성별'].replace('UNK', '여')
df_t['성별'].value_counts()
```

```python
train = df.iloc[:,1:-1]
target = df.iloc[:,-1]
```

```python
test_input = df_t.iloc[:,1:]
```

```python
from sklearn.preprocessing import OneHotEncoder
cols = train.select_dtypes("object").columns.tolist()
enc = OneHotEncoder(handle_unknown = 'ignore')
print(cols)
print(enc)
```

```python
# 학습데이터
tmp = pd.DataFrame(
    enc.fit_transform(train[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
train = pd.concat([train,tmp],axis=1).drop(columns=cols) # 범주형 컬럼 제거
train.head()
```

```python
# 테스트데이터
tmp = pd.DataFrame(
    enc.transform(test_input[cols]).toarray(), # 테스트데이터는 transform 만 해야한다.
    columns = enc.get_feature_names_out()
)
test_input = pd.concat([test_input,tmp],axis=1).drop(columns=cols) # 범주형 컬럼 제거
test_input.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# 학습데이터
train[train.columns] = scaler.fit_transform(train)
train.head()
```

```python
# test
test_input[test_input.columns] = scaler.transform(test_input)
test_input.head()
```

```python
#from sklearn.model_selection import train_test_split
#train_input, val_input, train_target, val_target = train_test_split(train,target,test_size=0.1,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
cv = KFold(n_splits=5,shuffle=True, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=17)
scores = cross_val_score(model,train,target,cv = cv ,scoring='f1',n_jobs = -1)
np.mean(scores)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(train,target)
```

```python
pred = model.predict(test_input)
pred.shape
```
