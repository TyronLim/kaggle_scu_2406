```python
# train에 갑상선저하인지, 갑상선항진증인지 합쳐서 [정상,갑상선비정상] 으로 나누기
def determine_thyroid_status(row):
    if row['갑상선저하_인지_여부'] == "예":
        return 'hypo'
    elif row['갑상선항진증_인지_여부'] == '예':
        return 'hyper'
    else:
        return 'Normal'

# 새로운 컬럼 추가
train['갑상선_인지'] = train.apply(determine_thyroid_status, axis=1)

# test에 갑상선저하인지, 갑상선항진증인지 합쳐서 [정상,갑상선비정상] 으로 나누기
def determine_thyroid_status(row):
    if row['갑상선저하_인지_여부'] == '예':
        return 'hypo'
    elif row['갑상선항진증_인지_여부'] == '예':
        return 'hyper'
    else:
        return 'Normal'

# 새로운 컬럼 추가
test['갑상선_인지'] = test.apply(determine_thyroid_status, axis=1)
```

```python
train['Pregnancy_Sex'] = train.apply(lambda row: '임신_여성' if row['임신_여부'] == '예' and row['성별'] == '여' else '기타', axis=1)
test['Pregnancy_Sex'] = test.apply(lambda row: '임신_여성' if row['임신_여부'] == '예' and row['성별'] == '여' else '기타', axis=1)
```

```python
x_train = train.drop(columns=['ID','target'])
y_train = train['target']
x_test = test.drop(columns=['ID'])
print(x_train.columns)
print(x_test.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
cols = x_train.select_dtypes("object").columns.tolist()
enc = OneHotEncoder(handle_unknown = 'ignore')
```

```python
# 학습데이터
tmp = pd.DataFrame(
    enc.fit_transform(x_train[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
train_ft = pd.concat([x_train,tmp],axis=1).drop(columns=cols) # 범주형 컬럼 제거
train_ft.head()
```

```python
# 테스트데이터
tmp = pd.DataFrame(
    enc.transform(x_test[cols]).toarray(), # 테스트데이터는 transform 만 해야한다.
    columns = enc.get_feature_names_out()
)
test_ft = pd.concat([x_test,tmp],axis=1).drop(columns=cols) # 범주형 컬럼 제거
test_ft.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# 학습데이터
train_ft[train_ft.columns] = scaler.fit_transform(train_ft)
train_ft.head()
```

```python
# 테스트 데이터
test_ft[test_ft.columns] = scaler.transform(test_ft)  # 테스트데이터는 transform 만 해야한다.
test_ft.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
cv = KFold(n_splits=5,shuffle=True, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(random_state=42,loss='exponential',
                                                    n_estimators=130,
                                                    learning_rate=0.1,
                                                    max_depth=3,
                                                    min_samples_split=6,
                                                    min_samples_leaf=2,
                                                    subsample=0.9,
                                                    max_features='sqrt')
scores = cross_val_score(model,train_ft,target,cv = cv ,scoring='f1',n_jobs = -1)
np.mean(scores)
```
